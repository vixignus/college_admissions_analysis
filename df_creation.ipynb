```python
def extract_marks(url):
    r = requests.get(url)
    bs = BeautifulSoup(r.text)
    tbl = bs.find("table", attrs={'class':'table table-striped'})
    for t in tbl.findAll('tr'):
        if 'th' in str(t):
            dept = t.text
        if 'td' in str(t):
            cc, marks = [m.text for m in t.findAll('td')]
            if marks == '-NIL-':
                marks = 0.0
            else:
                marks == float(marks)
            yield (dept, cc, marks)
```

```python
marks_df = pd.DataFrame()
for i, contents in df_colleges.iterrows():
    print(contents['link'], contents['name'])
    for dept, cc, marks in extract_marks(contents['link']):
        marks_df = marks_df.append(pd.Series([contents['link'], contents['name'],\
                                              dept, cc, marks]), \
                                   ignore_index=True)
```
