```python
import pandas as pd
import numpy as np
```

```python
marks_df = pd.read_pickle('marks_df.pkl')
colleges_df = pd.read_pickle('college_df.pkl')
```

```python
# merge the 2 datasets
golden_data = pd.merge(marks_df, colleges_df, on=['link', 'name'])
golden_data.head(5)
```

```python
depts = sorted(list(golden_data.dept.unique()))
circuital = ['electronics', 'computer science',\
             'information technology',\
             'instrumentation and control', 'nano',\
             'robotics']
mechanical = ['aeronautical', 'automobile', \
              'ceramic','chemical','civil',\
              'industrial engineering',\
              'manufacturing','marine',\
              'material', 'mechanical',\
              'mechatronics', 'metallurgical',\
              'petro', 'production','mining',\
              'polymer', 'rubber and plastic']
dept_stream_mapping = {}
for dept in depts:
    mapped_flag = 0
    # check for circuital
    for i in circuital:
        if i in dept.lower():
            dept_stream_mapping[dept] = "CIRCUITAL"
            mapped_flag = 1
            break
    # check for mechanical
    for i in mechanical:
        if i in dept.lower() or i == "CH & ELECT.CHEM(SS),KARAIKUDI":
            dept_stream_mapping[dept] = "MECHANICAL"
            mapped_flag = 1
            break
    # everything else
    if mapped_flag == 1:
        continue
    else:
        dept_stream_mapping[dept] = "OTHERS"
```

```python
dept_stream_mapping_df = pd.DataFrame(list(dept_stream_mapping.items()))
```

```python
dept_stream_mapping_df.columns = ['dept', 'stream']
```

```python
dept_stream_mapping_df[dept_stream_mapping_df.stream == 'OTHERS']
```

```python
golden_data = pd.merge(dept_stream_mapping_df, golden_data, on=['dept'])
```

```python
golden_data.head()
```

```python
marks_by_stream = pd.pivot_table(data=golden_data[(golden_data.cc == "OC")], \
               columns=['stream'], aggfunc='mean', \
               fill_value=0, values=['marks'], index=['name'])
```

```python
marks_by_stream.unstack().reset_index()
```
